```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Create a figure for the banner (16:9 aspect ratio)
fig = plt.figure(figsize=(16, 9))
plt.axis('off') # Hide axes

# Define title
fig.text(0.5, 0.92, "StayManager: SQL & Python Hotel Analytics", 
         ha='center', va='center', fontsize=28, fontweight='bold', color='#2c3e50')
fig.text(0.5, 0.86, "Dynamic Pricing | Availability Logic | Revenue Management", 
         ha='center', va='center', fontsize=18, color='#7f8c8d')

# Load the images you generated earlier
img1 = mpimg.imread('fig_1.png') # Revenue
img2 = mpimg.imread('fig_2.png') # Room Types
img3 = mpimg.imread('fig_3.png') # Nationalities
# If you have a 4th image (Occupancy), load it. If not, we use 3.

# Create Subplots for the collage
ax1 = fig.add_axes([0.05, 0.45, 0.40, 0.35]) # Top Left
ax1.imshow(img1)
ax1.axis('off')
ax1.set_title("Revenue Trends", fontsize=14)

ax2 = fig.add_axes([0.55, 0.45, 0.40, 0.35]) # Top Right
ax2.imshow(img2)
ax2.axis('off')
ax2.set_title("Room Popularity", fontsize=14)

ax3 = fig.add_axes([0.30, 0.05, 0.40, 0.35]) # Bottom Center
ax3.imshow(img3)
ax3.axis('off')
ax3.set_title("Guest Demographics", fontsize=14)

# Save the banner
plt.savefig('project_banner.png', dpi=150, bbox_inches='tight')
plt.show()
print("Project Cover Photo Saved as 'project_banner.png'")
```
